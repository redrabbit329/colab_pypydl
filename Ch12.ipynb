12.1. 선 그래프
12.1.1. 마커 종류와 색상 설정하기
   - matplotlib.pyplot.plot()을 이용하여 꺾은선 그래프를 그릴 수 있다.
   - marker = "마커 종류", markerfacecolor = "마커 색상"

```
# 12-1
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.arange(1,11)
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])

# 그래프를 설정합니다.
plt.ylim([0,weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

# 검은색의 원형 마ㅓ를 적용한 꺾은선 그래프를 작성하세요
plt.plot(days, weight, marker="o", markerfacecolor="k")
plt.show()
```

12.1.2. 선 종류와 색상 설정하기
   - matplotlib.pyplot.plot()에 데이터와 함께 linestyle="선종류", color="선색상"설정
   - 실선"-", 파선"--", 1점쇄선"-.", 점선:"

```
# 12-3
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.arange(1,11)
print(days)
weight = np.array([10,14,18,20,18,16,17,18,20,17])

#그래프프 설정합니다.
plt.ylim([0, weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

# 검은 색의 원형 마커를 설정하고, 파란색 파선을 적용한 꺾은 선 그래프를 작성하세요
plt.plot(days, weight, linestyle="--",color="b", marker="o", markerfacecolor="k")
plt.show()
```

12.2.2. 가로축에 라벨 설정하기
   -막대그래프에서는 matplotlib.pyplot.bar()에 tick_label로 라벨이 포함된 리스트를 전달하여 가로축 라벨을 설정할 수 있다

```
# 12-7
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5,6]
y = [12,41,32,36,21,17]
labels = ["Apple","Orange","Banana","Pineapple","Kiwifruit","Strawberry"]

# 막대그래프를 만들고, 가로축에 라벨을 설정
plt.bar(x,y, tick_label = labels)
plt.show()
```

12.2.3. 누적 막대그래프 만들기
   - matplotlib.pyplot.bar()의 bottom에 데이터 열(두 번째 이후, 플롯에서 지금까지의 합)을 전달하면, 해당 인덱스의 아래쪽 항목으로 누적 막대그래프를 설정 가능 
   - plt.legend(("y1", "y2")) 로 계열에 라벨을 지정할 수 있음

```
# 12-9
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5,6]
y1 = [12,41,32,36,21,17]
y2 = [43,1,6,17,17,9]
labels = ["Apple","Oranage","Banana","Pineapple","Kiwifruit","Strawberry"]

# 누적 막대그래프를 만들고 가로축에 라벨을 설정
plt.bar(x, y1, tick_label=labels)
plt.bar(x,y2,bottom=y1)

# 계열에 라벨을 지정
plt.legend(("y1","y2"))
plt.show()
```

12.6. 3D그래프
12.6.1. 3ㅇ Axes만들기
   - 3D 그래프를 만드려면, 3D랜더링 기능을 가진 서브플롯작성 필요함
   - projection="3d"로 3D 렌더링 기능을 가진 서브플롯을 작성할 수 있다.
   - matplotlib.figure().add_subplot(1,1,1,projection="3d")

```
# 12-35
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t,t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1,projection="3d")
ax.plot_surface(X,Y,Z)
plt.show()
```

12.6.2.곡면 만들기

```
# 12-37
# 서브 플롯 ax.plot_surface(X,Y,Z)
import numpy as np
import matplotlib.pyplot as plt
# 3D 랜더링에 필요한라이브러리
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

x = y = np.linspace(-5,5)
X,Y = np.meshgrid(x,y)
Z = np.exp(-(X**2 + Y**2)/2) / (2*np.pi)

# figure객체 생성
fig = plt.figure(figsize=(6,6))
# 서브플롯 ax를 만듬
ax = fig.add_subplot(1,1,1,projection="3d")
# 곡면을 만듬
ax.plot_surface(X,Y,Z)
plt.show()
```

```
# 12-43
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)
X,Y = np.meshgrid(t,t)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1,projection="3d")

ax.plot_surface(X,Y,Z, cmap=cm.coolwarm)
plt.show()
```

```
# 12-47 종합문제
# 목표 : 난수 생성 방법, 데이터를 그래프에 플롯하는 방법, 몬테카를로 방법
import matplotlib.pyplot as plt
import numpy as np
import math
import time

%matplotlib inline

np.random.seed(100)

# 원 안에 들어간 횟수
X = 0

# 시도 횟수 N을 지정하세요
N = 1000

# 사분면 경계의 방정식을 그립니다.
circle_x = np.arange(0, 1, 0.001)
circle_y = np.sqrt(1-circle_x * circle_x)
plt.figure(figsize=(5,5))
plt.plot(circle_x, circle_y)

# N번 시도에 걸리는 시간을 측정합니다.
start_time = time.clock()

# N번 시도합니다.
for i in range(0,N):

  # 0~1사이의 균등 난수를 생성하여 변수 score_x에 저장하세요
  score_x = np.random.rand()

  # 0~1사이의 균등 난수를 생성하여 변수 score_y에 저장하세요
  score_y = np.random.rand()

  # 점이 원 안에 들어간 경우와 들어가지 않은 경우를 조건 분기하세요
  if score_x*score_x + score_y*score_y <1:

    # 원 안에 들어가면 검은색으로, 들어가지 않으면 파란색으로 표시하세요
    plt.scatter(score_x, score_y, marker='o', color='k')

    # 원 안에 들어갔다면 위에서 정의한 변수 X에 1포인트를 더하세요
    X = X + 1
  else:
    plt.scatter(score_x, score_y, marker='o', color='b')

# pi의 근사값을 계산하세요
pi = 4*float(X)/float(N)

# 몬테카를로 방법의 실행 시간을 계산
end_time = time.clock()
time = end_time - start_time

# 원주율의 결과를 표시하세요
print("원주율 : %.6f" % pi)
print("실행시간 : %f" % (time))

print("실행시간 : %f" %(time))

# 결과를 표시합니다
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
